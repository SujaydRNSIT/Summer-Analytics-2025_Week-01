### Find the car (or cars) with the best ratio of horsepower to weight among all cars that also have above-median mpg.

```python
#your code here
# Import pandas
import pandas as pd

# Read the CSV file
df = pd.read_csv('C:/Users/ALL/Desktop/SA2025_Week-1/Cars.csv')

# Convert 'horsepower' to numeric, handling missing values
df['horsepower'] = pd.to_numeric(df['horsepower'], errors='coerce')

# Calculate horsepower-to-weight ratio
df['hp_per_weight'] = df['horsepower'] / df['weight']

# Calculate median MPG
median_mpg = df['mpg'].median()

# Filter cars with above-median MPG
above_median_mpg = df[df['mpg'] > median_mpg]

# Find the maximum horsepower-to-weight ratio
max_ratio = above_median_mpg['hp_per_weight'].max()

# Get the car(s) with the maximum ratio
best_ratio_cars = above_median_mpg[above_median_mpg['hp_per_weight'] == max_ratio][['name', 'mpg', 'horsepower', 'weight', 'hp_per_weight']]

# Print results
print(f"Median MPG: {median_mpg}")
print(f"Maximum horsepower-to-weight ratio: {max_ratio}")
print("Car(s) with the best ratio:")
print(best_ratio_cars)
```

### Design a multi-line plot using Matplotlib or Seaborn that shows the evolution of average mpg over the years, separately for each origin

```python
#your code here
# Import required libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Set Seaborn style for better aesthetics
sns.set_style("whitegrid")

# Read the CSV file
df = pd.read_csv('C:/Users/ALL/Desktop/SA2025_Week-1/Cars.csv')

# Verify data
print("Missing values:\n", df[['mpg', 'model_year', 'origin']].isnull().sum())
print("Unique origins:", df['origin'].unique())

# Group by model_year and origin, calculate mean mpg
avg_mpg = df.groupby(['model_year', 'origin'])['mpg'].mean().reset_index()

# Create the multi-line plot
plt.figure(figsize=(10, 6))
sns.lineplot(data=avg_mpg, x='model_year', y='mpg', hue='origin', 
             style='origin', markers=True, palette='deep')

# Customize the plot
plt.xlabel('Model Year')
plt.ylabel('Average MPG')
plt.title('Evolution of Average MPG by Origin (1970–1982)')
plt.legend(title='Origin', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()

# Save the plot
plt.savefig('mpg_evolution.png')
plt.show()
```

### We define a “consistent” car model as one that was produced over multiple years and had very low variation in mpg across those years (standard deviation < 1.0).

Tasks:

Identify car names that appear in more than one model_year.

For each such name, compute the standard deviation of mpg across years.

Return the car(s) with the lowest variation in mpg, among those with at least 2 appearances and std(mpg) < 1.0.

Report the model name(s), number of appearances, and the average mpg.

Bonus:
Sort the result by number of appearances (descending), then mpg (descending).

```python
#your code here
# Import pandas
import pandas as pd

# Read the CSV file
df = pd.read_csv('C:/Users/ALL/Desktop/SA2025_Week-1/Cars.csv')

# Verify data
print("Missing MPG values:", df['mpg'].isnull().sum())
print("Total cars:", df.shape[0])

# Group by 'name' to find cars with multiple model years
car_counts = df.groupby('name')['model_year'].nunique()

# Filter cars with more than one model_year
multi_year_cars = car_counts[car_counts > 1].index

# Filter DataFrame for these cars
multi_year_df = df[df['name'].isin(multi_year_cars)]

# Group by 'name' to compute stats
stats = multi_year_df.groupby('name').agg(
    appearances=('model_year', 'nunique'),
    mpg_std=('mpg', 'std'),
    mpg_mean=('mpg', 'mean')
)

# Filter for cars with mpg standard deviation < 1.0
consistent_cars = stats[(stats['mpg_std'] < 1.0) & (stats['appearances'] >= 2)]

# Sort by appearances (descending) and mpg_mean (descending)
consistent_cars = consistent_cars.sort_values(by=['appearances', 'mpg_mean'], ascending=[False, False])

# Print results
print("Consistent cars (std(mpg) < 1.0, appearances >= 2):")
print(consistent_cars[['appearances', 'mpg_mean', 'mpg_std']])
```
